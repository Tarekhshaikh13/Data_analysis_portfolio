### convert salary into integer + separate col for min and max salary

```python
import re
# Define a function to extract min and max salary range from the string
def extract_salary_range(salary_str):
    # Check if the string contains "Per Hour"
    if "Per Hour" in salary_str:
        matches = re.findall(r'\$(\d+)', salary_str)
        if len(matches) == 2:
            min_salary = int(matches[0])*40*52
            max_salary = int(matches[1])*40*52
            return [min_salary, max_salary]
    else:
        # Use regular expressions to find min and max values in K
        matches = re.findall(r'\$(\d+)K', salary_str)
        if len(matches) == 2:
            min_salary = int(matches[0]) * 1000
            max_salary = int(matches[1]) * 1000
            return [min_salary, max_salary]
   

# Apply the function to the 'SalaryEstimate' column
dataset['Salary Estimate']= dataset['Salary Estimate'].apply(extract_salary_range)
# Salary
```

```python
#new data frame of salary range
salary_range = pd.DataFrame(dataset['Salary Estimate'].tolist(), columns = ["min_salary","max_salary"])

#insert above dataframe into orignal datafram
dataset = pd.concat([dataset.iloc[:,:3],salary_range,dataset.iloc[:,3:]],axis=1 )

#drop orignal salary column
dataset = dataset.drop(columns=["Salary Estimate"])

dataset
```

### minimum and maximum revenue cloumns in millions

```python
dataset['Revenue'].value_counts()
```

```python
Revenue = dataset['Revenue']
Revenue = Revenue.str.replace('$','').str.replace('(USD)','').str.replace(' ','').str.replace('+billion','to11billion')
Revenue = Revenue.replace('Unknown/Non-Applicable',np.nan)
Revenue = Revenue.str.replace('billion','000')
Revenue = Revenue.str.replace('million','').str.replace('Lessthan','0to')
Revenue = Revenue.str.split('to',expand=True)
# test = Revenue[Revenue=='1to5']
# test.str.split('to',expand=True).value_counts()
# Revenue.iloc[:,0][Revenue.iloc[:,0]=='1'].value_counts()
Revenue.columns = ["min_revenue_in_millions","max_revenue_in_millions"]

Revenue = Revenue.applymap(lambda x :  int(x) if pd.notna(x) and x.isdigit() else x)
Revenue
```

```python
df = pd.concat([dataset.iloc[:,:-3],Revenue,dataset.iloc[:,-3:]],axis=1)
dataset = df
dataset = dataset.drop(columns=['Revenue'])
dataset
```
